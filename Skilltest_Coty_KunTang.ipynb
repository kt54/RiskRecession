```python
#3
grouped4 = df2.groupby('Geography').sum()
grouped4['avg_sale_per_unit'] = grouped4['Net Sales']/grouped4['Volume']
grouped4['trade_funding_level'] = (grouped4['Gross Sales']-grouped4['Net Sales'])/grouped4['Gross Sales']
```

```python
#3a)
max_dis = grouped4['avg_sale_per_unit'].max()-grouped4['avg_sale_per_unit'].min()
print('The biggest discrepancy is',max_dis,'between',grouped4['avg_sale_per_unit'].idxmax(),'and',grouped4['avg_sale_per_unit'].idxmin())
```

```python
grouped4
```

```python
trad_fund = grouped4.sort_values(by='trade_funding_level',ascending=False)[:10]
```

```python
#3（b)
trad_fund.plot.pie(y=['trade_funding_level'],figsize=(15,15))
plt.show()
```
